# Create Player Dataset
This is a notebook that creates a player dataset by scraping Eliteprostects using Marcus Sjölins scraper eliteprospect_scraper.

```python
# Import the eliteprospect_scraper made by Marcus Sjölin
import eliteprospect.eliteprospect_scraper as ep
import pandas as pd
import time
```

```python
# Use data from 10 least seasons
seasons = ['2014-2015', '2015-2016', '2016-2017', '2017-2018', '2018-2019', '2019-2020', '2020-2021', '2021-2022', '2022-2023', '2023-2024']

# Get player links for all players that have played in the SHL during the last 10 seasons
all_links = set()
for season in seasons:
    # Get players playing in SHL the specific season
    df_players = ep.getPlayers('shl', season)
    # Add EP player liks to a set
    links = set(df_players['link'])
    # Update the set with the new links
    all_links.update(links)

# Convert the set to a DataFrame with the column name 'player_links'
df = pd.DataFrame(all_links, columns=['player_links'])

# Write the DataFrame to a CSV file
df.to_csv("player_links.csv", index=False)

# Leagues
leagues = ['shl', 'allsvenskan', 'division-1', 'liiga', 'nla', 'del', 'czech', 'nhl', 'ahl', 'echl', 'ncaa']
```
